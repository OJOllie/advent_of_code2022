```python
#Day 2 , Part 1

import pandas as pd
import numpy as np 

# Read in data with a column for game result (e.g AY)

df = pd.read_csv("day2.txt", sep=" ",names=["Opponent", "Response"])
df["Result"] = df["Opponent"] + df["Response"]

# Assign values for initial throw and game result

points = {}


points[1] = ['X'] #rock A
points[2] = ['Y'] #paper B
points[3] = ['Z'] #scissors C

result = {}

result[6] = ['AY','BZ','CX']
result[3] = ['AX','BY','CZ']
result[0] = ['AZ', 'BX', 'CY']

# Create new columns calling to dictionary values for throw and game result

def f(s):
    for k in points:
        for v in points[k]:
            if v in s:
                return k

df["ThrowPoints"] = df["Response"].apply(f)

def f(s):
    for k in result:
        for v in result[k]:
            if v in s:
                return k

df["ResultPoints"] = df["Result"].apply(f)

# Sum value and return score

df['OverallPoints'] = df['ThrowPoints'] + df['ResultPoints']
Total = df['OverallPoints'].sum()
print(Total)
```

```python
# Part 2

df = pd.read_csv("day2.txt", sep=" ",names=["Opponent", "Response"])
df["Result"] = df["Opponent"] + df["Response"]

points = {}

points[1] = ['X'] #rock A
points[2] = ['Y'] #paper B
points[3] = ['Z'] #scissors C

requiredresult = {}

requiredresult[0] = ['X'] # required loss
requiredresult[3] = ['Y'] # required draw
requiredresult[6] = ['Z'] # required win

yourthrow = {}

yourthrow['X'] = ['AY','BX','CZ']
yourthrow['Y'] = ['AZ','BY','CX']
yourthrow['Z'] = ['AX','BZ','CY']
     
# Create new columns calling to dictionary values for throw and game result

def f(s):
    for k in requiredresult:
        for v in requiredresult[k]:
            if v in s:
                return k

df["ResultPoints"] = df["Response"].apply(f)

            
def f(s):
    for k in yourthrow:
        for v in yourthrow[k]:
            if v in s:
                return k

df["YourThrow"] = df["Result"].apply(f)

def f(s):
    for k in points:
        for v in points[k]:
            if v in s:
                return k

df["ThrowPoints"] = df["YourThrow"].apply(f)

# Sum value and return score

df['OverallPoints'] = df['ResultPoints'] + df['ThrowPoints']
Total = df['OverallPoints'].sum()
print(Total)

```
